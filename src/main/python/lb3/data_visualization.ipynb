```python
import matplotlib.pyplot as plt

data = {
    1: [5.237581968307495, 5.2839391231536865, 5.216270208358765, 5.279177188873291, 5.269155979156494],
    2: [2.785627841949463, 2.8358352184295654, 2.851968288421631, 2.803236484527588, 2.80647611618042],
    3: [1.961272954940796, 2.027587652206421, 1.9361677169799805, 1.9766252040863037, 2.0969929695129395],
    6: [1.2627770900726318, 1.2486956119537354, 1.2163329124450684, 1.2226030826568604, 1.1776747703552246]
}

processes = []
avg_times = []

for p, times in data.items():
    processes.append(p)
    avg_times.append(sum(times) / len(times))

processes, avg_times = zip(*sorted(zip(processes, avg_times)))

plt.figure()
plt.plot(processes, avg_times, marker='o')

plt.xlabel('Число процессов')
plt.ylabel('Среднее время (с)')
plt.title('Зависимость среднего времени от числа процессов')
plt.grid()

plt.show()
```
