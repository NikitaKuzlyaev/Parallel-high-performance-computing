```
import matplotlib.pyplot as plt

data = {
    1024: [2072, 2042, 2018],
    8192: [2732, 2807, 2610],
    4096: [2264, 2147, 2220],
    2048: [2095, 2136, 2058],
    512: [1964, 2004, 1982],
    256: [2039, 1967, 1996],
    128: [1996, 1967, 2055],
    64: [1889, 1887, 1889],
    1: [10919, 10867, 10865],
    2: [5575, 5539, 5571],
    3: [3779, 3811, 3791],
    4: [2925, 2948, 3088],
    5: [2914, 2858, 2885],
    6: [2591, 2596, 2556],
    7: [2482, 2483, 2460],
    72: [1839, 1931, 1935],
    8: [2377, 2282, 2302],
    9: [2215, 2193, 2203],
    10: [2092, 1999, 2036],
    11: [1883, 1913, 1982],
    12: [1899, 1844, 1906],
    16: [1890, 1881, 1867],
    20: [1901, 1862, 1937],
    24: [1892, 1888, 1930],
    28: [1857, 1928, 1874],
    96: [1993, 2029, 2026],
    32: [1880, 1957, 1919],
    48: [1927, 1846, 1887],
}

avg_times = {k: sum(v) / len(v) for k, v in data.items()}

threads = sorted(avg_times.keys())
times = [avg_times[t] for t in threads]

plt.figure(figsize=(14, 6))

plt.plot(threads, times, marker='x')

plt.xscale("log", base=1.2)

plt.xticks(threads, threads, rotation=90)

plt.xlabel("Количество потоков")
plt.ylabel("Среднее время выполнения (мс)")
plt.title("Зависимость времени выполнения программы от количества потоков")

plt.grid(True, which="both")
plt.grid(axis='x', which='both', alpha=0.3)
plt.grid(axis='y', which='both', alpha=0.7)

plt.tight_layout()
plt.show()
```

```
avg_times = {k: sum(v) / len(v) for k, v in data.items()}

threads = sorted(avg_times.keys())

# базовое время (1 поток)
base_time = avg_times[1]

# ускорение
speedup = [base_time / avg_times[t] for t in threads]

plt.figure(figsize=(14, 6))

plt.plot(threads, speedup, marker='x')

plt.xscale("log", base=1.2)

plt.xticks(threads, threads, rotation=90)

plt.xlabel("Количество потоков")
plt.ylabel("Ускорение (раз)")
plt.title("Ускорение выполнения программы относительно одного потока")

plt.grid(axis='x', which='both', alpha=0.3)
plt.grid(axis='y', which='both', alpha=0.7)

plt.tight_layout()
plt.show()
```
